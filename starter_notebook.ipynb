### Foundation Data Processing Functions

````python
# 🤖 AI Collaboration: Data Loading and Cleaning Function
# Ask AI to help you create a robust data loading function

def load_and_clean_transaction_data(file_path):
    """
    Load and clean transaction data for the Smart Finance Assistant

    🤖 AI Collaboration Prompt:
    "Create a function to load CSV transaction data with Date, Amount, Category,
    Description columns. Handle dollar signs in Amount, missing values, and
    data validation. Include clear business-focused error messages."

    Args:
        file_path: Path to CSV file or file object
    Returns:
        pandas.DataFrame: Cleaned transaction data
    """
    import pandas as pd
    import numpy as np

    # Load data - handle both file paths and DataFrames
    if isinstance(file_path, pd.DataFrame):
        df = file_path.copy()
    else:
        try:
            df = pd.read_csv(file_path)
        except FileNotFoundError:
            raise ValueError(f"❌ File not found: {file_path}")
        except Exception as e:
            raise ValueError(f"❌ Error reading file: {str(e)}")

    # Validate required columns
    required_columns = ['Date', 'Amount', 'Category', 'Description']
    missing_columns = [col for col in required_columns if col not in df.columns]
    if missing_columns:
        raise ValueError(f"❌ Missing required columns: {', '.join(missing_columns)}")

    print(f"📥 Loaded {len(df)} transactions")

    # Clean Amount column
    print("\n🧹 Cleaning Amount data...")
    original_amounts = df['Amount'].copy()

    # Replace empty strings with NaN
    df['Amount'] = df['Amount'].replace('', np.nan)

    # Remove dollar signs and convert to float
    df['Amount'] = df['Amount'].astype(str).str.replace('$', '', regex=False)
    df['Amount'] = df['Amount'].str.replace(',', '', regex=False)  # Handle thousands separator
    df['Amount'] = pd.to_numeric(df['Amount'], errors='coerce')

    # Report missing values
    missing_count = df['Amount'].isna().sum()
    if missing_count > 0:
        print(f"⚠️  Warning: {missing_count} transactions have missing amounts (will be excluded)")
        # Remove rows with missing amounts
        df = df.dropna(subset=['Amount'])

    # Clean Date column
    print("\n📅 Cleaning Date data...")
    try:
        df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
        invalid_dates = df['Date'].isna().sum()
        if invalid_dates > 0:
            print(f"⚠️  Warning: {invalid_dates} invalid dates found (will be excluded)")
            df = df.dropna(subset=['Date'])
    except Exception as e:
        raise ValueError(f"❌ Error parsing dates: {str(e)}")

    # Sort by date
    df = df.sort_values('Date').reset_index(drop=True)

    # Clean Category and Description
    df['Category'] = df['Category'].str.strip()
    df['Description'] = df['Description'].str.strip()

    # Identify refunds and regular transactions
    refunds = df[df['Amount'] < 0]
    expenses = df[df['Amount'] > 0]

    print(f"\n✅ Data cleaning complete!")
    print(f"   • {len(df)} valid transactions")
    print(f"   • {len(expenses)} expenses (positive amounts)")
    print(f"   • {len(refunds)} refunds (negative amounts)")
    print(f"   • Date range: {df['Date'].min().date()} to {df['Date'].max().date()}")

    return df

# Test your function
print("🧪 Testing data cleaning function...")
print("="*80)

clean_data = load_and_clean_transaction_data(df_sample)

print("\n" + "="*80)
print("📊 Cleaned Data Preview:")
print(clean_data.head(10))

print("\n" + "="*80)
print("📈 Data Quality Summary:")
print(clean_data.info())

print("\n" + "="*80)
print("💰 Amount Statistics:")
print(clean_data['Amount'].describe())
````

````python
# 🤖 AI Collaboration: Spending Analysis Function
# Ask AI to help you create comprehensive spending analysis

def analyze_spending_patterns(df):
    """
    Analyze spending patterns and generate business insights

    🤖 AI Collaboration Prompt:
    "Create a function that analyzes spending by category, calculates percentages,
    identifies top spending areas, and generates actionable financial insights
    formatted for business presentation."

    Args:
        df: Cleaned transaction DataFrame
    Returns:
        dict: Analysis results and insights
    """
    import pandas as pd

def analyze_spending_patterns(df):
    """
    Analyzes spending patterns, calculates distribution, and generates
    actionable financial insights formatted for business presentation.
    """

    # 1. Aggregating spending by category
    category_totals = df.groupby('Category')['Amount'].sum().sort_values(ascending=False)
    total_spend = category_totals.sum()

    # 2. Calculating percentages
    category_percentages = (category_totals / total_spend * 100).round(2)

    # 3. Generating Actionable Financial Insights
    insights = []

    # Insight A: Concentration Risk
    top_cat = category_totals.index[0]
    top_pct = category_percentages.iloc[0]
    if top_pct > 45:
        insights.append(f"HIGH CONCENTRATION: {top_cat} accounts for {top_pct}% of total spend. Evaluate vendor dependency.")
    else:
        insights.append(f"HEALTHY MIX: Spending is diversified; {top_cat} is the lead driver at {top_pct}%")

    # Insight B: Tail Spend Analysis
    if len(category_totals) > 12:
        insights.append("PROCESS OPTIMIZATION: High number of small expense categories detected. Consider consolidating tail-spend vendors.")

    # 4. Formatting for Business Presentation
    analysis_results = {
        "Executive_Summary": {
            "Total_Expenditure": f"${total_spend:,.2f}",
            "Primary_Driver": top_cat,
            "Category_Count": len(category_totals)
        },
        "Spending_Breakdown": pd.DataFrame({
            "Total_Amount": category_totals,
            "Percentage_Share": category_percentages
        }).to_dict(orient='index'),
        "Top_3_Priorities": category_totals.head(3).index.tolist(),
        "Strategic_Recommendations": insights
    }

    return analysis_results

# --- Test the function ---
analysis = analyze_spending_patterns(clean_data)
import json
print("\n" + "="*80)
print("📊 Spending Analysis Results:")
print(json.dumps(analysis, indent=4))
````

````python
import pandas as pd
import gradio as gr

def load_and_clean_data(file_obj):
    try:
        df = pd.read_csv(file_obj.name)
        df.columns = df.columns.str.strip() # Remove hidden spaces

        # 1. DYNAMIC COLUMN DETECTION
        # Look for the 'Money' column
        amount_col = None
        for col in ['Amount', 'Total_Sale', 'Unit_Price']:
            if col in df.columns:
                amount_col = col
                break

        # Look for the 'Category' column
        category_col = None
        for col in ['Category', 'Product', 'Region']:
            if col in df.columns:
                category_col = col
                break

        if not amount_col or not category_col:
            return f"Error: Could not find required columns. Found: {list(df.columns)}"

        # 2. CLEANING THE MONEY COLUMN
        if df[amount_col].dtype == 'object':
            # Remove symbols like $, --, and commas
            df[amount_col] = df[amount_col].astype(str).str.replace('$', '', regex=False)
            df[amount_col] = df[amount_col].str.replace(',', '', regex=False).str.strip()
            df[amount_col] = pd.to_numeric(df[amount_col], errors='coerce')

        # Drop rows with empty amounts
        df = df.dropna(subset=[amount_col])

        return df, amount_col, category_col
    except Exception as e:
        return f"System Error: {str(e)}", None, None

def smart_analysis_interface(file):
    # Process the file
    data, amount_col, cat_col = load_and_clean_data(file)

    # If 'data' is a string, it means an error occurred
    if isinstance(data, str):
        return data

    # Run Analysis
    total = data[amount_col].sum()
    top_group = data.groupby(cat_col)[amount_col].sum().idxmax()
    breakdown = data.groupby(cat_col)[amount_col].sum().sort_values(ascending=False)

    # Format the English Report
    report = [
        "📊 DYNAMIC DATA ANALYSIS",
        "=" * 30,
        f"Detected Category Column: {cat_col}",
        f"Detected Money Column: {amount_col}",
        "-" * 30,
        f"Total Value: ${total:,.2f}",
        f"Top Performing {cat_col}: {top_group}",
        "\nDETAILED BREAKDOWN:",
    ]

    for name, value in breakdown.items():
        report.append(f"• {name}: ${value:,.2f}")

    return "\n".join(report)

# Launch the Gradio App
demo = gr.Interface(
    fn=smart_analysis_interface,
    inputs=gr.File(label="Upload any CSV file"),
    outputs=gr.Textbox(label="Analysis Report", lines=15),
    title="Universal Finance & Sales Assistant"
)

demo.launch()
# Launch your complete application
# demo = create_finance_assistant_ui()
# demo.launch()
````
